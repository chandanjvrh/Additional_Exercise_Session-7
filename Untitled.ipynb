```python
#Import iris dataset
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
data=pd.DataFrame(iris.data,columns=iris.feature_names)
label=pd.DataFrame(list(map(lambda x : iris.target_names[x],iris.target)),columns=['Species'])
iris=pd.concat([data,label],axis=1)
print(iris.head())
```

1)Use distplot() to see the distribution of the SapelLengthCm,SapelWidthCm,PetalLengthCm,PetalWidthCm . 
  Plot them as subplots in a single image

```python
import seaborn as sb
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10,10), ncols=2, nrows=2)
sb.distplot(data['sepal length (cm)'],ax=ax[0,0])
sb.distplot(data['sepal width (cm)'],ax=ax[0,1])
sb.distplot(data['petal length (cm)'],ax=ax[1,0])
sb.distplot(data['petal width (cm)'],ax=ax[1,1])
```

```python
#Do a box plot of all the features except 'Species'
fig = plt.figure(figsize=(20,7))
sb.boxplot(data=data)
```

```python
# Do a count plot for the feature 'Species'
fig, ax = plt.subplots(figsize=(7, 3))
sb.countplot(y=iris['Species'], data=iris, color="b");
```

```python
# Do a pair plot on features sepal length, sepal width,petal length,petal width, Species

sb.pairplot(iris,hue='Species',size=3)
```

```python
#Do an Implot of the following SepalLengthCm ,PetalLengthCm .Using hue. display the different species in different
# Colors.

sb.lmplot(x="sepal length (cm)", y="petal length (cm)", hue="Species", data=iris)
```

```python
# Do a bar plot of 'Species' vs 'SepalLengthcm'

sb.barplot(x='sepal length (cm)',y='Species' , data=iris )
```

```python
# Using heatmap , plot correlation matrix.

sb.heatmap(iris.corr(),cmap='coolwarm')
plt.title('Iris correlation')
```
